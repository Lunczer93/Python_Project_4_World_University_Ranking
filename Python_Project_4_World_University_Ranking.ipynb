##### Selecting and renaming columns, then data wrangling & cleaning 

```python
NewWorldUniversityRanking20172022 = WorldUniversityRanking20172022.iloc[:,[0,1,2,3,5,6,7,9,10,11,12,13,14]]
NewWorldUniversityRanking20172022.rename(columns={
                                                   'university':'University',
                                                   'year':'Year',
                                                   'rank_display':'Rank',
                                                   'score':'Score',
                                                   'country':'Country',
                                                   'city':'City',
                                                   'region':'Region',
                                                   'type':'Type',
                                                   'research_output':'Research Output',
                                                   'student_faculty_ratio':'Student Faculty Ratio',
                                                   'international_students':'International Students',
                                                   'size':'Size',
                                                   'faculty_count':'Faculty Count'}, inplace=True)
NewWorldUniversityRanking20172022
```

```python
for col in NewWorldUniversityRanking20172022.columns:
    pct_missing = round(np.mean(NewWorldUniversityRanking20172022[col].isnull()),2)
    print('{}-{}%'.format(col, pct_missing))
```

```python
NewWorldUniversityRanking20172022[NewWorldUniversityRanking20172022.isna().any(axis=1)]
```

```python
NewWorldUniversityRanking20172022['Rank'].fillna(0,inplace=True)
NewWorldUniversityRanking20172022['Score'].fillna(0,inplace=True)
NewWorldUniversityRanking20172022['City'].fillna('No Data',inplace=True)
NewWorldUniversityRanking20172022['Type'].fillna('No Data',inplace=True)
NewWorldUniversityRanking20172022['Research Output'].fillna('No Data',inplace=True)  
NewWorldUniversityRanking20172022['Student Faculty Ratio'].fillna(0,inplace=True)
NewWorldUniversityRanking20172022['International Students'].fillna(0,inplace=True)
NewWorldUniversityRanking20172022['Size'].fillna('No Data',inplace=True)
NewWorldUniversityRanking20172022['Faculty Count'].fillna(0,inplace=True)
```

```python
(NewWorldUniversityRanking20172022.isnull().sum()/(len(NewWorldUniversityRanking20172022)))*100
```

```python
NewWorldUniversityRanking20172022.duplicated().sum()
```

```python
NewWorldUniversityRanking20172022.dtypes
```

```python
NewWorldUniversityRanking20172022['International Students'] =NewWorldUniversityRanking20172022['International Students']\
                                                            .str.replace(',', '.').astype(float)
                                                                    

NewWorldUniversityRanking20172022['Faculty Count'] =NewWorldUniversityRanking20172022['Faculty Count']\
                                                            .str.replace(',', '.').astype(float)

```
